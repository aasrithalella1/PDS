```python
df = pd.read_csv("Used_car_data.csv")
df.head()
```

#A) Look for the missing values in all the columns and imputed them with Median

```python
# Checks how many missing values are in each column
df.isnull().sum()
```

Removed text units and converted Mileage, Engine, and Power to numeric.

```python
# Removes units and converts to numeric before filling missing values
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '', regex=False)
df['Mileage'] = df['Mileage'].str.replace(' km/kg', '', regex=False)
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')

df['Engine'] = df['Engine'].str.replace(' CC', '', regex=False)
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')

df['Power'] = df['Power'].str.replace(' bhp', '', regex=False)
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')

# Fills missing values
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
df['Power'] = df['Power'].fillna(df['Power'].median())
df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])


# Dropped the 'New_Price' column since it has too many missing values
df = df.drop(columns=['New_Price'])
```

Filled missing values using median (for numeric) and mode (for categorical).

```python
print("Missing values handled")
print(df.isnull().sum())
```

#B) Remove the units from some of the attributes and only keep the numerical values

Cleaned text units like "kmpl", "CC", "bhp" to keep only numerical values.

```python
#Removes units from numerical columns

# Removes ' kmpl' and ' km/kg' from Mileage
df['Mileage'] = df['Mileage'].astype(str).str.replace(' kmpl', '', regex=False)
df['Mileage'] = df['Mileage'].str.replace('km/kg', '', regex=False)

# Remove ' CC' from Engine
df['Engine'] = df['Engine'].astype(str).str.replace(' CC', '', regex=False)

# Remove ' bhp' from Power
df['Power'] = df['Power'].astype(str).str.replace(' bhp', '', regex=False)

# Converts all to numeric
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')
```

#C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

Converted categorical values into binary columns using one-hot encoding

```python
#Onehot encode Fuel_Type and Transmission
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'], drop_first=True)
```

```python
print("Onehot encoding done, new columns:")
print(df.columns)
```

#E)Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset

Filtered rows where the car is older than 5 years.

```python
#car age is greater than 5 years
df_filtered = df[df['Car_Age'] > 5]

print(df_filtered.head())
```

Renamed columns to clearer names.

```python
# Rename columns
df_renamed = df.rename(columns={'Car_Age': 'Age_of_Car', 'Price': 'Car_Price'})

print(df_renamed.head())
```

```python
# Mutate
df['Log_Price'] = df['Price'].apply(lambda x: np.log(x) if x > 0 else 0)

print(df[['Price', 'Log_Price']].head())
```

```python
# Group by both 'Fuel_Type_Electric' and 'Fuel_Type_Petrol', and calculate the average 'Price'
df_grouped = df.groupby(['Fuel_Type_Electric', 'Fuel_Type_Petrol'])['Price'].mean().reset_index()

print(df_grouped)
```
